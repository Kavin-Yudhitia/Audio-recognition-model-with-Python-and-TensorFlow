```python
import pandas as pd

test_dataset_df = pd.read_csv(filepath_or_buffer="test_dataset_df.csv", index_col=0)
prediction_result_df = pd.read_csv(filepath_or_buffer="prediction_result.csv", index_col=0)
```

```python
result_join_df = test_dataset_df.merge(prediction_result_df, on='split', how='left').drop(['split'], axis=1)
result_join_df = result_join_df.sort_values(['file', 'start'], ascending=[True, True])
print(index)
result_join_df.head(10)
```

```python
# result_join_df = pd.read_csv(filepath_or_buffer="time stamp.csv", index_col=0).reset_index() # just for testing

category_dict = {'category': []}

category_count = 0
category_current = -1
file_current = ""

for index, row in result_join_df.iterrows():
    if (category_current != row['water']) or (file_current != row['file']):
        category_count += 1
        category_current = row['water']
        file_current = row['file']
    category_dict['category'].append(category_count)
    
result_join_df = result_join_df.join(pd.DataFrame.from_dict(category_dict))
result_join_df.head(40)
```

```python
final_df = result_join_df.groupby('category').agg(file=('file', 'max'), start=('start', 'min'), stop=('stop', 'max'), water=('water', 'min')).reset_index().drop(columns = ['category'], axis = 1)
final_df.head(40)
```

```python
final_df.to_csv('final_result.csv')
```
